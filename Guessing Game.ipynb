## Project Description

* **`First: You ask the user to set the game level:`**
  * *(1) Easy*: 
    * Limits : [1 - 10]  
    * No. of trials : 3
  * *(2) Intermediate*: 
    * Limits : [1 - 100] 
    * No. of trials : 7
  * *(3) Hard* :  
    * Limits : [1 - 1000] 
    * No. of trials : 15

* **`Second: start the game according to the game level:`**
  * set the `guess` to a number within the limits using `random` module 
  * set the number of trials: `n_trials` 

* **`Finally: start the game:`**
  * Ask the user to guess the number:
    * If he guessed it successfully : `print('Congratulations, you achieved it in {user_trials} trial') `
    * if he failed:
      * user_trials < n_trials: give him another trial within
      * user_trials = n_trials: `print('You Lose!')`

# Template of Solution

## Show the game levels

```python
def show_levels():
  print("""You can choose Your Game Level:
        (1) Easy: 
              Limits : [1 - 10]  
              No. of trials : 3
        (2) Intermediate: 
              Limits : [1 - 100] 
              No. of trials : 7
        (3) Hard :  
              Limits : [1 - 1000] 
              No. of trials : 15""")

show_levels()

  # write your code here 
```

## Ask the user for the game level

```python
def game_level_choice():
   while True:
      global game_level
      game_level = int(input(f"Please Choose Game Level :\n (1) Easy \n (2) Intermediate \n (3) Hard \n"))
      if game_level in (1,2,3):
        break
      else:
        print(f"Error Please Choose form options!")
        continue  
   return game_level          
game_level = game_level_choice()
```

## Set the game settings according to the game level:


```python
print(game_level)
def set_game_settings(game_level):
  if game_level == 1:
    limits = 10
    n_trials = 3
  elif game_level == 2:
    limits = 100
    n_trials = 7 
  else:
    limits = 1000
    n_trials = 15
  return limits, n_trials

limits, n_trials = set_game_settings(game_level)
print(limits)
print(n_trials)
```

## Start Playing

> Hint: to generate random number in some range:

`import random`

`num = random.randint(lower, upper)`

 or

 `num = random.choice(range(lower, upper))`


```python
import random
def start_play(limits, n_trials):
  tried_trails=1
  num = random.randint(1, limits)  
  while True:
   
   
      UserNumber = int(input(f"Please Enter Number Between 1 and {limits} ! This Is try number {tried_trails} of {n_trials}\n"))
      
      if(UserNumber==num):
        print(f"You tried the correct number")
        break
      else:
        if(tried_trails==n_trials):
          print(f"Sorry You tried {n_trials} times and the correct number is {num}")
          break
        elif(UserNumber>num):
          print(f"You entered Wrong Number Please Decreese It")
          tried_trails=tried_trails+1
          continue
        elif(UserNumber<num):
          print(f"You entered Wrong Number Please Increase It")
          tried_trails=tried_trails+1
          continue

start_play(limits, n_trials)



```

## Let's Play

```python
def play():
  show_levels()
  game_level = game_level_choice()
  limits, n_trials = set_game_settings(game_level)
  start_play(limits, n_trials)
```

```python
play()
```
